###### In this project I will try to answer the question: 'Does a country's total GDP have an impact on life expectancy in that country?' 

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('all_data.csv')
print(df.head())
```

```python
def Plot_GDP_Life(df,country):
    '''This function takes a datafile and a string as arguments. It creates a temporary datafile
    using the passed string argument and creates a figure with two subplots. It plots life expectancy
    and GDP of the country vs year.'''
    
    new_df = df[df.Country == country]
    fig,ax = plt.subplots(1,2,figsize=(13,6))
    
    ax[0].plot(new_df.Year, new_df['Life expectancy at birth (years)'],marker='o')
    ax[0].set_xlabel("Year")
    ax[0].set_ylabel("Life Expectancy")
    ax[0].set_title('Life Expectancy per Year in {}'.format(country))
    ax[0].grid()

    ax[1].plot(new_df.Year, new_df['GDP'],marker='o',color='green')
    ax[1].set_xlabel("Year")
    ax[1].set_ylabel("GDP")
    ax[1].set_yticklabels(['0','$100B','$150B','$200B','$250B','$300B','$350B','$400B','$450B','$500B'])
    ax[1].set_title('GPD per Year in {}'.format(country))
    ax[1].grid()

    plt.show()
    plt.close('all')
    
```

```python
Plot_GDP_Life(df,'Chile')
```

```python
Plot_GDP_Life(df,'China')
```

```python
Plot_GDP_Life(df,'Germany')
```

```python
Plot_GDP_Life(df,'Mexico')
```

```python
Plot_GDP_Life(df,'United States of America')
```

```python
Plot_GDP_Life(df,'Zimbabwe')
```

- From the graphs of each country listed above, it can be assumed that a country's GDP does have a bearing on the life expectancy of that country. The graphs follow a similar path indicating one effects the other.
